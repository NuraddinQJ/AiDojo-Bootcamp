# Bar Chars

```python
cat = ['A', 'B', 'C', 'D']
val = [10, 15 , 12 ,8]
x = np.arange(len(cat))

plt.bar(x, val, 0.2)
plt.xticks(x,cat)

plt.show()
```

```python
cat = ['A', 'B', 'C', 'D']
val = [10, 15 , 12 ,8]
x = np.arange(len(cat))

plt.barh(x, val, 0.4)
plt.yticks(x,cat)

plt.show()
```

```python
cat = ['A', 'B', 'C', 'D']
val1 = [31, 15 , 12 ,8]
val2 = [17, 11, 0 ,29]
val3 = [13, 45, 12, 34]
w = 0.2

x = np.arange(len(cat))
plt.bar(x, val1, w)
plt.bar(x+w, val2, w)
plt.bar(x-w, val3, w)

plt.xticks(x,cat)
plt.show()
```

```python
import gradio as gr
```

```python
def make_array(a,b,stp):
  return str(np.arange(a,b,stp))

with gr.Blocks() as demo:
  with gr.Column():
    a = gr.Number()
    b = gr.Number()
    stp = gr.Number()
    btn = gr.Button("Run")


  with gr.Column():
    out= gr.Text()

  btn.click(fn=make_array, inputs=[a,b,stp], outputs=[out])

demo.launch(share=True)
```

```python
import pandas as pd
```

```python
arra = []
```

```python
def creat_df(a,b,stp):
  global arra
  arra = pd.DataFrame({"array":np.arange(a,b,stp)})
  return arra

def opporation(scaler, oper, arr):
  if oper == "Add":
    arr = arra + scaler
  elif oper == "Sub":
    arr = arra - scaler
  elif oper == "Mult":
    arr = arra * scaler
  elif oper == "Div":
    arr = arra / scaler

  return arr


with gr.Blocks() as demo:
  with gr.Tab("Create Array:"):
    with gr.Row():
      with gr.Column():
        a = gr.Number()
        b = gr.Number()
        stp = gr.Number()
        btn = gr.Button("Run")

      with gr.Column():
        out= gr.DataFrame()

    btn.click(fn=creat_df, inputs=[a,b,stp], outputs=[out])

  with gr.Tab("Opearation:"):
    with gr.Row():
      with gr.Column():
        scl = gr.Number(value=1, label="Scaler")
        radio = gr.Radio(["Add", "Sub", "Mult", "Div"], value="Add", label="Chose Operassion")
        btn1 = gr.Button("Run")

      with gr.Column():
        out2 = gr.DataFrame(label="")

      btn1.click(fn=opporation,inputs=[scl,radio, out],outputs=[out2])


demo.launch()
```

```python
def plot(df , radio):
  x = df['x'].values.astype(np.int32)
  y = df['y'].values.astype(np.int32)

  if radio == "plot":
    plt.title("Line Chart")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.plot(x,y)

  return plt

with gr.Blocks() as demo:
  with gr.Column():

    df = gr.DataFrame(label="Input Dataframe", headers=["x", "y"], max_rows=5, max_cols=2)
    radio = gr.Radio(["plot", "hist", "bar"])
    btn = gr.Button("Show")
    out = gr.Plot(label="My Image")

  btn.click(fn=plot, inputs=[df, radio], outputs=[out])

demo.launch()
```
