```python
# Show the Plots for the following columns
# Pie chart for the Survived ,Sex,Embarked and Pclass
plt.figure(figsize=(16, 16))
plt.subplot(3,4,1)
plt.pie(data.Survived.value_counts(), labels=[0, 1], autopct='%1.f%%')
plt.title('Survived')
plt.subplot(3,4,2)
plt.pie(data.Pclass.value_counts(), labels=[1, 2 ,3], autopct='%1.f%%')
plt.title('Pclass')
plt.subplot(3,4,3)
plt.pie(data.Sex.value_counts(), labels=['Male', 'Female'], autopct='%1.f%%')
plt.title('Sex')
plt.subplot(3,4,4)
plt.pie(data.Embarked.value_counts(), labels=['S', 'C', 'Q'], autopct='%1.f%%')
plt.title('Embarked')


# Bar chart for the SibSp and Parch
plt.figure(figsize=(16, 12))
plt.subplot(3,4,6)
plt.bar(data.SibSp.value_counts().index, data.SibSp.value_counts())
plt.title('SibSP')
plt.subplot(3,4,7)
plt.bar(data.Parch.value_counts().index, data.Parch.value_counts())
plt.title('Parch')
# Histogram for the Age and Fare
plt.figure(figsize=(16, 12))
plt.subplot(3,4,9)
plt.hist(data.Age.dropna(), bins=10)
plt.title('Age')
plt.subplot(3,4,10)
plt.hist(data.Fare.dropna(), bins=10)
plt.title('Fare')
plt.show()

```
